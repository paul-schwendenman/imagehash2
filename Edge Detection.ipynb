# Create helper functions

create and demo some filters. make bunch of helper for playing with settings

```python
def find_edges(image):
    out = image.filter(ImageFilter.FIND_EDGES)
    imshow(out)
    return out
find_edges(image1)
```

```python
def resize(image, hash_size=8, antialias=True):
    if antialias:
        out = image.resize((hash_size, hash_size), Image.ANTIALIAS)
    else:
        out = image.resize((hash_size, hash_size))
    imshow(out)
    return out

resize(image1)
```

```python
resize(image1, 64, False)
```

```python
def grayscale(image):
    out = image.convert('L')
    imshow(out)
    return out
imshow(grayscale(image1).convert('RGB'))
```

```python
def bwshow(image):
    imshow(image, interpolation='none', cmap=cm.gray)
bwshow(grayscale(image1))
```

```python
def mode(image, value=25):
    out = image.filter(ImageFilter.ModeFilter(value))
    imshow(out)
    return out
image1_mode = mode(image1)
```

```python
find_edges(image1_mode)
```

```python
def median(image, value=5):
    out = image.filter(ImageFilter.MedianFilter(value))
    imshow(out)
    return out
median(image1)
```

```python
def min_filter(image, value=5):
    out = image.filter(ImageFilter.MinFilter(value))
    imshow(out)
    return out
image1_min = min_filter(image1, 9)
```

```python
find_edges(image1_min)
```

```python
def max_filter(image, value=5):
    out = image.filter(ImageFilter.MaxFilter(value))
    imshow(out)
    return out
image1_max = max_filter(image1, 9)
```

```python
def blur(image, value=5):
    out = image.filter(ImageFilter.GaussianBlur(value))
    imshow(out)
    return out
image1_blur = blur(image1, 9)
```

```python
min_filter(image1_blur)
```

```python
find_edges(min_filter(image1_blur))
```

```python
def sharpen(image):
    out = image.filter(ImageFilter.SHARPEN)
    imshow(out)
    return out
sharpen(min_filter(image1_blur))
```

```python
def contour(image):
    out = image.filter(ImageFilter.CONTOUR)
    imshow(out)
    return out
contour(image1)
```
